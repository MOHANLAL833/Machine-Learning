```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the data
data=pd.read_csv("50_Startups.csv")
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data.columns
```

```python
independent=data[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent=data[["Profit"]]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#https://scikit-learn.org/1.5/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='squared_error', splitter='random')
regressor=regressor.fit(xtrain,ytrain)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(ytest,y_pred)
r_score
```
